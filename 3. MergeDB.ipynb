```python
date_id = pd.read_csv("/content/drive/MyDrive/MASTER EN DATA SCIENCE/Capstone/data_dsmarket/GrupoRetail_1/Proceso Total/BBDD output/date_id_2.csv")
```

```python
date_id.head()
```

```python
sales_calendar = pd.merge(
    left = sales_calendar,
    right = date_id,
    how = 'left',
    on = ["id", "year_week"])
```

```python
sales_calendar.head()
```

```python
prices = pd.read_csv("/content/drive/MyDrive/MASTER EN DATA SCIENCE/Capstone/data_dsmarket/GrupoRetail_1/Proceso Total/BBDD input/item_prices.csv")
```

```python
prices["yearweek2"] = prices['yearweek'].astype(str)
prices["year"] = prices['yearweek2'].str[:4]
prices["week"] = prices["yearweek2"].str[4:6]
prices["year_week"] = prices["year"] + "-" + prices["week"]
```

```python
prices.head()
```

```python
prices["id"] = prices["item"] + "_" + prices["store_code"]
```

```python
prices.drop(["yearweek", "yearweek2"], axis=1, inplace=True)
```

```python
prices.head()
```

```python
prices.drop(["item", "category", "store_code"], axis=1, inplace = True)
```

```python
sales_calendar_prices = pd.merge(
    left = sales_calendar,
    right = prices,
    how = "left",
    on = ["id", "year_week" ]
)
```

```python
sales_calendar_prices.drop(["year_week"], axis=1, inplace = True)
```

```python
sales_calendar_prices.head()
```

```python
database = sales_calendar_prices.copy()
```

```python
database["sell_price"] = database["sell_price"].round(2)
```

```python
database.to_csv("/content/drive/MyDrive/MASTER EN DATA SCIENCE/Capstone/data_dsmarket/GrupoRetail_1/Proceso Total/BBDD output/database_v5.csv", index=False, header=True, decimal=',', sep=';', float_format='%.3f')
```

```python
gb = database.groupby(["id", "item", "category", "department", "store", "region", "year"]).sum()
```

```python
gb.drop("sell_price", axis=1, inplace=True)
```

```python
gb.reset_index(inplace=True)
```
